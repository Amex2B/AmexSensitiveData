```python
# created a new dataframe 'new_df' since I didn't want to mess with the original
# only taking the first 50 rows from the source_text of the original dataset so I can manually go 
# through and check the correctness of the labelling or if there is any errors
phones2_df = pd.DataFrame(train_df.loc[0:50, 'source_text'])
phones2_df
```

```python
# regex pattern that looks for phone numbers in a variety of formats
#pattern = r'/(\+?\d{1,3}[-.\s]?)?\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}/gm' didn't work because of the /gm at the end??
pattern = r'(\+?\d{1,3}[-.\s]?)?\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}'

# checking robustness of regex code for numbers
#trial_df is ONLY a trial data set made up by me
data = {'trial_set': ['+1 1234567890','+11234567890','1234567890','123-456-7890','(123)456-7890','(123)4567890','(123)456-7890',
                      '+1(123)4567890','+1 123 456 7890','123 456 7890','+1 (123)-4567890','+1 (123)-456-7890','(123)-456-7890',
                      '(123)-(456)-(7890)','blah blah\n\n1234567890\n\n','blahblah1234567890blahblah',':::];/.1234567890,.;',
                      'nothing to see here', 'just text and some number 12312', '1232321 not a phone number']}
trial_df=pd.DataFrame(data)
trial_df['has_phone_number'] = trial_df['trial_set'].str.contains(pattern)
trial_df
```

```python
#creating a new column in the new_df that is supposed to be the 'label' that the model is meant to be trained on
phones2_df['has_phone_number'] = phones2_df['source_text'].str.contains(pattern)
phones2_df
```

```python
#found another dataset with mostly names
from datasets import load_dataset
dataset = load_dataset("ai4privacy/pii-masking-65k")
names = dataset['train']  
names_df = names.to_pandas()
names_df.columns
```

```python
# regex pattern that looks for names (only if they are capitalized and separated with a space in between)
# this has a lot of problems 
#e.g. "Hello Bob" is considered the same as "Bob Dylan" -- both will be classified as names
pattern = r'\b[A-Z][a-zA-Z\'-]*\b \b[A-Z][a-zA-Z\'-]*\b'

#needs contextual information for singular names that aren't capitialized but could ignore 
#this requirement for now, to just test the model

#decided to make my own dataset, with a column (1 or 0 representing true or false if it has a name)
# however, I realize that the goal is to redact the information, not just identify that it 
# exists in a bunch of text

#nevermind, it wouldn't really work for the task at hand, it's best to tokenize

#to test
names2_df = pd.DataFrame(names_df.loc[0:50, 'unmasked_text'])
names2_df['has_name'] = names2_df['unmasked_text'].str.contains(pattern)
names2_df
```

```python
# using list comprehension + split()
# Tokenizing strings in list of strings
res = [sub.split() for sub in names2_df['unmasked_text']]
names2_df['tokenized_names'] = res
names2_df
```

```python
from datasets import load_dataset
import pandas as pd

ds = load_dataset("Isotonic/pii-masking-200k")
```

```python
def extract_account_numbers_english(dataset):
    account_data = []

    # Iterate over each example in the dataset
    for example in dataset:
        # Check if the language is English (assuming 'en' for English)
        if example['language'] == 'en':
            # Convert the privacy_mask string back into a dictionary
            privacy_mask = eval(example['privacy_mask'])

            # Check if '[ACCOUNTNUMBER_1]' exists in the privacy_mask
            account_entries = {key: value for key, value in privacy_mask.items() if '[ACCOUNTNUMBER_1]' in key}

            # If we have any '[ACCOUNTNUMBER_1]' entries, extract them
            if account_entries:
                # Iterate over all ACCOUNTNUMBER_1 entries (could be one or more)
                for account_key, account_value in account_entries.items():
                    # Add to our account_data list, each ACCOUNT entry as a new row
                    account_data.append({
                        'unmasked_text': example['unmasked_text'],
                        'masked_text': example['masked_text'],
                        'account_number': account_value
                    })

    return account_data

# Extract account numbers from the train set, only for English rows
account_data = extract_account_numbers_english(ds['train'])

# Convert extracted data to a DataFrame for saving
df_account = pd.DataFrame(account_data)

# Save the DataFrame to a CSV file
output_csv_account = 'extracted_account_numbers_english.csv'
df_account.to_csv(output_csv_account, index=False)

print(f"Extracted account numbers from English rows saved to {output_csv_account}")
```

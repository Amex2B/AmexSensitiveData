```python
import random
# Select 100 random indices
random_indices = random.sample(range(len(train_set)), 100)

# Get the random samples
random_samples = train_set.select(random_indices)

print(random_samples)
```

```python
# Load model directly
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("lakshyakh93/deberta_finetuned_pii")
```

```python
# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples['source_text'], truncation=True)

encoded_tr = random_samples.map(tokenize_function, batched = True)

print(encoded_tr)
print(encoded_tr['input_ids'][:10])
```

```python
from transformers import AutoModelForTokenClassification, AutoModelForPreTraining # could use this one later

model = AutoModelForTokenClassification.from_pretrained("lakshyakh93/deberta_finetuned_pii")
#text = "My name is John and I live in California."
#output = model(text, aggregation_strategy="first")
#print(output)

from transformers import pipeline
gen = pipeline("token-classification", "lakshyakh93/deberta_finetuned_pii", device=-1)

text = "My name is John and I live in California."
print(gen(text, aggregation_strategy="first"))
```

```python
from transformers import TrainingArguments

#training_args = TrainingArguments(output_dir="test_trainer", eval_strategy="epoch")
train_args = TrainingArguments("test_trainer")
```

```python
import evaluate
metric=evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=1)
    results = metric.compute(references=labels, predictions=predictions)
```

```python
from transformers import Trainer

small_val=val_set.shuffle(seed=123).select(range(100))
encoded_small_val=small_val.map(tokenize_function, batched=True)
```

```python
trainer = Trainer(model,
                  train_args,
                  train_dataset=encoded_tr,
                  eval_dataset=encoded_small_val,
                  tokenizer=tokenizer,
                  compute_metrics=compute_metrics)
```

```python
trainer.train()
```

```python
trainer.evaluate()
```
